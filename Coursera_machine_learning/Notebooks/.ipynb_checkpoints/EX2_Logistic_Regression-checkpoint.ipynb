```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pd.read_csv(filepath_or_buffer='../Data/ex2data1.txt', header = None, names = ['Exam1','Exam2','Accepted'])
data1.head()
```

```python
plt.figure(figsize=[8,8])
plt.scatter(x=data1.where(data1.Accepted==1).Exam1.values,y=data1.where(data1.Accepted==1).Exam2.values,marker='+',label='Accepted')
plt.scatter(x=data1.where(data1.Accepted==0).Exam1.values,y=data1.where(data1.Accepted==0).Exam2.values,marker='o',label='Reject')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend(loc=3)
plt.show()
```

```python
def costFunction(X,y,thetas):
    m = y.size
    h = lambda z: 1/(1+np.exp(-z))
    z = X@thetas
    J = (1/m)*(-(y.T@np.log(h(z)))-((1-y).T@np.log(1-h(z))))
    return J[0]

def gradient_descent(X, y, alpha = 0.01, iterations = 1000):
    m = y.size
    num_j = X.shape[1]
    thetas = np.zeros((num_j,1))
    new_thetas = np.zeros_like(thetas)
    h = lambda z: 1/(1+np.exp(-z))
    J_list = []
    for _ in range(iterations):
        for j in range(num_j):
            grad = (h(X@thetas) - y).T@np.c_[X[:,j]]
            new_thetas[j] = thetas[j] - alpha*grad[0,0]/m
        J_list.append(costFunction(X,y,thetas))
        thetas = new_thetas
        #print(thetas.T)
    return thetas, J_list
```

```python
X = data1.values.copy()
y = np.c_[X[:,2]]
X = np.c_[np.ones((X.shape[0],1)), X[:,:-1]]
```

```python
thetas, J_list = gradient_descent(X,y,alpha=.0021, iterations=1000000)
```

## So this is why the problem recommends an alternate method of optimization
1,000,000 iterations... really? And going any higher with alpha causes the wobbles in the first 10,000 iterations to be even worse.

It took me so long to realize it wasn't actually converging in the first 100 iterations, because it certainly looks like that.

```python
plt.plot(J_list)
plt.show()
```

```python
x1, x2 = np.meshgrid(np.linspace(0,100),np.linspace(0,100))
grid_points = np.c_[np.ones((len(x1.flatten()),1)),x1.flatten(),x2.flatten()]
values = 1/(1+np.exp(-grid_points@thetas))
```

```python
plt.figure(figsize=[8,8])
plt.scatter(x=data1.where(data1.Accepted==1).Exam1.values,y=data1.where(data1.Accepted==1).Exam2.values,marker='+',label='Accepted')
plt.scatter(x=data1.where(data1.Accepted==0).Exam1.values,y=data1.where(data1.Accepted==0).Exam2.values,marker='o',label='Reject')
plt.contour(x1,x2,values.reshape(50,50), levels = [.5])
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.xlim(.95*X[:,1].min(), 1.02*X[:,1].max())
plt.ylim(.95*X[:,2].min(), 1.02*X[:,2].max())
plt.legend(loc=3)
plt.show()
```

Now I want to know if normalizing the initial data would help me.

```python
X = data1.values.copy()
y = np.c_[X[:,2]]
X = np.c_[np.ones((X.shape[0],1)), X[:,:-1]/100] # Only difference is dividing by 100 here
thetas, J_list = gradient_descent(X,y,alpha=30, iterations=50)
plt.plot(J_list)
plt.xlabel('# of Iterations')
plt.ylabel('Cost')
plt.show()

x1, x2 = np.meshgrid(np.linspace(0,1),np.linspace(0,1))
grid_points = np.c_[np.ones((len(x1.flatten()),1)),x1.flatten(),x2.flatten()]
values = 1/(1+np.exp(-grid_points@thetas))

plt.figure(figsize=[8,8])
plt.scatter(x=data1.where(data1.Accepted==1).Exam1.values/100,y=data1.where(data1.Accepted==1).Exam2.values/100,marker='+',label='Accepted')
plt.scatter(x=data1.where(data1.Accepted==0).Exam1.values/100,y=data1.where(data1.Accepted==0).Exam2.values/100,marker='o',label='Reject')
plt.contour(x1,x2,values.reshape(50,50), levels = [.5])
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.xlim(.95*X[:,1].min(), 1.02*X[:,1].max())
plt.ylim(.95*X[:,2].min(), 1.02*X[:,2].max())
plt.legend(loc=3)
plt.show()
```

So yeah, putting the data in a space more similar to the classification value makes it converge A LOT faster, and allows me to crank up the alpha value much more easily.

```python
data2 = pd.read_csv(filepath_or_buffer='../Data/ex2data2.txt', header = None, names = ['Test1','Test2','Good'])
data2.head()
```

```python
plt.figure(figsize=[8,8])
plt.scatter(x=data2.where(data2.Good==1).Test1.values,y=data2.where(data2.Good==1).Test2.values,marker='+',label='Good')
plt.scatter(x=data2.where(data2.Good==0).Test1.values,y=data2.where(data2.Good==0).Test2.values,marker='o',label='Bad')
plt.xlabel('Test1')
plt.ylabel('Test2')
plt.legend(loc=3)
plt.show()
```

```python
X = data2.values.copy()
y = np.c_[X[:,2]]
def extraParameterize(X):
    X_new = np.ones_like(X)
    X_new = np.c_[np.ones((X.shape[0],1)),\
                  X[:,0],\
                  X[:,1],\
                  X[:,0]**2,\
                  X[:,0]*(X[:,1]),\
                  X[:,1]**2,\
                  X[:,0]**3,\
                 (X[:,0]**2)*(X[:,1]),\
                 (X[:,0])*(X[:,1]**2),\
                  X[:,1]**3,\
                  X[:,0]**4,\
                 (X[:,0]**3)*(X[:,1]),\
                 (X[:,0]**2)*(X[:,1]**2),\
                 (X[:,0])*(X[:,1]**3),\
                  X[:,1]**4,\
                  X[:,0]**5,\
                 (X[:,0]**4)*(X[:,1]),\
                 (X[:,0]**3)*(X[:,1]**2),\
                 (X[:,0]**2)*(X[:,1]**3),\
                 (X[:,0])*(X[:,1]**4),\
                  X[:,1]**5,\
                  X[:,0]**6,\
                 (X[:,0]**5)*(X[:,1]),\
                 (X[:,0]**4)*(X[:,1]**2),\
                 (X[:,0]**3)*(X[:,1]**3),\
                 (X[:,0]**2)*(X[:,1]**4),\
                 (X[:,0])*(X[:,1]**5),\
                  X[:,1]**6]
    return X_new
X = extraParameterize(X)
```

```python
def Normalization(X):
    X_new = np.ones_like(X)
    n = X.shape[1]
    mean_list, max_list, min_list = ([] for _ in range(3))
    for i in range(1,n):
        vec = X[:,i]
        cur_mean = vec.mean()
        mean_list.append(cur_mean)
        cur_max = vec.max()
        max_list.append(cur_max)
        cur_min = vec.min()
        min_list.append(cur_min)
        X_new[:,i] = (vec-cur_mean)/(cur_max-cur_min)
    return X_new, mean_list, max_list, min_list
```

```python
X, mean_list, max_list, min_list = Normalization(X)
```

## Now to add regularization to the cost function and gadient descent

```python
def costFunctionRegularization(X,y,thetas, lamda): #intentional misspelling
    m = y.size
    h = lambda z: 1/(1+np.exp(-z))
    z = X@thetas
    J = (1/m)*(-(y.T@np.log(h(z)))-((1-y).T@np.log(1-h(z)))) + (lamda/(2*m))*thetas.T@thetas
    return J[0]

def gradientDescentRegularization(X, y, alpha = 0.01, lamda = 1, iterations = 1000):
    m = y.size
    num_j = X.shape[1]
    thetas = np.zeros((num_j,1))
    new_thetas = np.zeros_like(thetas)
    h = lambda z: 1/(1+np.exp(-z))
    J_list = []
    for _ in range(iterations):
        for j in range(num_j):
            grad = (h(X@thetas) - y).T@np.c_[X[:,j]] + lamda*thetas[j]*(j>=1)
            new_thetas[j] = thetas[j] - alpha*grad[0,0]/m
        J_list.append(costFunctionRegularization(X,y,thetas,lamda))
        thetas = new_thetas
    return thetas, J_list
```

```python
thetas, J_list = gradientDescentRegularization(X,y,alpha=5, lamda = .0001, iterations=1000)
plt.plot(J_list)
plt.xlabel('# of Iterations')
plt.ylabel('Cost')
plt.show()
```

```python
x1, x2 = np.meshgrid(np.linspace(-1,1,100),np.linspace(-1,1,100).T)
grid_points = np.c_[x1.flatten(),x2.flatten()]
grid_points = extraParameterize(grid_points)
for i in range(len(mean_list)):
    grid_points[:,i+1] = (grid_points[:,i+1]-mean_list[i])/(max_list[i]-min_list[i])
values = 1/(1+np.exp(-grid_points@thetas))
```

```python
plt.figure(figsize=[8,8])
plt.scatter(x=X[y.ravel()==1,1],y=X[y.ravel()==1,2],marker='+',label='Good')
plt.scatter(x=X[y.ravel()==0,1],y=X[y.ravel()==0,2],marker='+',label='Bad')
plt.contour((x1-mean_list[0])/(max_list[0]-min_list[0]),(x2-mean_list[1])/(max_list[1]-min_list[1]),values.reshape(100,100), levels = [.5])
plt.xlabel('Test1')
plt.ylabel('Test2')
plt.legend(loc=3)
plt.show()
```
