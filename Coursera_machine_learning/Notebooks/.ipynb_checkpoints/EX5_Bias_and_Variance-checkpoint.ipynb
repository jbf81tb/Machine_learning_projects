```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
%matplotlib inline
```

```python
data = loadmat('../Data/ex5data1.mat')
```

```python
data.keys()
```

The data describes change in water levels (X) and the amount of water leaving a dam (y). It is split into training data, cross-validation data (\_val) and testing data (\_test).

```python
X = data['X']
y = data['y']
Xval = data['Xval']
yval = data['yval']
Xtest = data['Xtest']
ytest = data['ytest']
```

```python
X     = np.c_[np.ones((X.shape[0],1)),     X]
Xval  = np.c_[np.ones((Xval.shape[0],1)),  Xval]
Xtest = np.c_[np.ones((Xtest.shape[0],1)), Xtest]
```

```python
plt.scatter(X,y)
plt.xlabel('Change in water level')
plt.ylabel('Water flowing out of the dam')
plt.show()
```

It might work best to just do it with raw code, instead of the sklearn implementation, so I'll start with that. Most of the work was done in EX2 already, just have to change it from logistic to linear.

```python
def costFunctionRegularization(X,y,thetas, lamda): #intentional misspelling
    #Expect X to be bias-offset. (i.e. X = [1, X])
    m = X.shape[0]
    J = (1/m)*np.sum((X@thetas-y)**2) + (lamda/(2*m))*np.sum(thetas[1:,:]**2)
    return J

def gradientDescentRegularization(X, y, alpha = 0.01, lamda = 1, iterations = 100):
    m = X.shape[0]
    num_j = X.shape[1]
    thetas = np.zeros((num_j,1))
    J_list = []
    for _ in range(iterations):
        grad = X.T@(X@thetas - y) + lamda*thetas*(np.c_[np.arange(num_j)]>=1)
        thetas -= alpha*grad/m
        J_list.append(costFunctionRegularization(X,y,thetas,lamda))
    return thetas, J_list
```

```python
thetas,J_list = gradientDescentRegularization(X,y,alpha = .002,iterations=2000, lamda=0)
```

```python
plt.plot(J_list)
plt.show()
```

Because I don't want to bother regularizing the dataset in this very manual methodology, it's taking a bit longer to converge. But at least it still runs nearly instantaneously.

```python
plt.scatter(X[:,1],y,c='k', label='data')
plt.plot(X[:,1],X@thetas,c='r',label='fit')
plt.legend()
plt.xlabel('Change in water level')
plt.ylabel('Amount of water leaving the dam')
plt.show()
```

For a linear fit, the regularization wasn't useful, but I anticipate adding some polynomial features since right now we have high bias.

## Making some learning curves

```python
import time
from IPython.display import clear_output
```

```python
J_train_curve, J_val_curve = ([],[])
curve_count = range(X.shape[1]+1,X.shape[0])
for i in curve_count:
    clear_output(wait=True)
    thetas,J_list = gradientDescentRegularization(X[:(i+1),:].reshape(-1,2),y[:(i+1)].reshape(-1,1),alpha = .0016,iterations=2000, lamda=0)
    plt.plot(J_list)
    plt.title(i)
    plt.show()
    J_train_curve.append(J_list[-1])
    J_val_curve.append(costFunctionRegularization(Xval,yval.reshape(-1,1),thetas,lamda=0))
    #time.sleep(1/2)
```

```python
plt.plot(curve_count,J_train_curve,c='r',label='train')
plt.plot(curve_count,J_val_curve,c='k',label='validation')
plt.legend()
plt.ylim((0,100))
plt.title('Learning Curve')
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.show()
```

So it looks like our validation and training data don't even get that close with a full set of data, and they don't appear to be getting any closer with more data. Even more reason to believe that the data is biased/underfit.

```python
def poly_features(X,p,need_renorm=True, renorms = []):
    #X should be an (m,2) matrix where the first column is all 1's and the second is the relavent data
    newX = np.copy(X)
    for i in range(2,p):
        newX = np.c_[newX, np.reshape(newX[:,1]**i,(-1,1))]
    if need_renorm:
        mu = np.mean(newX[:,1:],axis=0)
        sigma = np.max(newX[:,1:],axis=0) - np.min(newX[:,1:],axis=0)
    else:
        mu = renorms[0]
        sigma = renorms[1]
    newX[:,1:] = (newX[:,1:]-mu)/sigma
    return newX, [mu, sigma]
```

Since the polynomial feature adding have high powers, we have to renormalize the values so it doesn't blow up. I'm sure there's a better way to deal with these things than I currently am.

```python
ymu = y.mean()
ysigma = y.max()-y.min()
ynorm = (y-ymu)/ysigma
```

```python
min_p, max_p = (2,13)
power_list = range(min_p,max_p)
lam_list = [0, .001, .003, .01, .03, .1, .3, 1, 3, 10]
J_train_list = np.zeros((len(power_list),len(lam_list)))
J_val_list = np.zeros((len(power_list),len(lam_list)))
for ip, p in enumerate(power_list):
    for ilam, lam in enumerate(lam_list):
        newX, renorms = poly_features(X,p)
        clear_output(wait=True)
        thetas, J_list = gradientDescentRegularization(newX,ynorm,alpha=1,lamda=lam,iterations=1000)
        J_train_list[ip,ilam] = costFunctionRegularization(newX,ynorm,thetas=thetas,lamda=0)
        plt.plot(J_list)
        plt.title(f'power = {p}, lambda = {lam}')
        plt.show()
        newValX, _ = poly_features(Xval,p,need_renorm=False,renorms=renorms)
        J_val_list[ip,ilam] = costFunctionRegularization(newValX,(yval-ymu)/ysigma,thetas=thetas,lamda=0)
    
```

```python
plt.figure(figsize=[8,8])
plt.imshow(np.log(J_val_list), cmap=plt.get_cmap('plasma'))
plt.colorbar()
plt.xticks(range(len(lam_list)),[str(lam) for lam in lam_list])
plt.xlabel('Lambda')
plt.yticks(range(len(power_list)),[str(p-1) for p in power_list])
plt.ylabel('Power')
plt.title('Log of cost')
plt.show()
```

```python
divmod(J_val_list.argmin(),J_val_list.shape[1])
```

```python
print(f'The optimal polynomial to use is to the {power_list[6]-1}th power with a regularization of {lam_list[6]}')
```

Obviously, I could zoom in on this region to try to find something _more_ optimal, but not right now.

```python
J_train_curve, J_val_curve, J_test_curve = ([],[],[])
newX, renorms = poly_features(X,power_list[6])
curve_count = range(np.intp(np.ceil(np.sqrt(newX.shape[1]))),X.shape[0]+1)
for i in curve_count:
    clear_output(wait=True)
    thetas,J_list = gradientDescentRegularization(newX[:i,:],ynorm[:i].reshape(-1,1),alpha=1,iterations=1000, lamda=lam_list[6])
    plt.plot(J_list)
    plt.title(i)
    plt.show()
    J_train_curve.append(costFunctionRegularization(newX[:(i+1),:],ynorm[:(i+1)].reshape(-1,1),thetas,lamda=0))
    newXval, _ = poly_features(Xval,power_list[6],need_renorm=False,renorms=renorms)
    J_val_curve.append(costFunctionRegularization(newXval,(yval-ymu)/ysigma,thetas,lamda=0))
    newXtest, _ = poly_features(Xtest,power_list[6],need_renorm=False,renorms=renorms)
    J_test_curve.append(costFunctionRegularization(newXtest,(ytest-ymu)/ysigma,thetas,lamda=0))
    #time.sleep(.2)
```

```python
plt.plot(curve_count,J_train_curve,c='r',label='train')
plt.plot(curve_count,J_val_curve,c='k',label='validation')
plt.plot(curve_count,J_test_curve,c='g',label='test data')
plt.legend()
#plt.ylim((0,100))
plt.title('Learning Curve')
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.show()
```

It certainly looks like the train and validation converge pretty well. And it looks like the test data behaves the same as the validation data, so that's good.

However, when dealing with an 8th degree polynomial it feels like we don't really have enough data with only 12 examples to work with.
